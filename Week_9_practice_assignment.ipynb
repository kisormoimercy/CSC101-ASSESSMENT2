````python
import pandas as pd
import numpy as np

# ---------------------------------------------------------
# 1. LOAD + CLEAN DATA
# ---------------------------------------------------------
df = pd.read_csv("transactions.csv")

# Clean Amount column
df["Amount"] = (
    df["Amount"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)

df["Date"] = pd.to_datetime(df["Date"])
df["Category"] = df["Category"].str.strip()

# ---------------------------------------------------------
# 2. MONTHLY TOTALS BY CATEGORY
# ---------------------------------------------------------
monthly_totals = (
    df.groupby("Category")["Amount"]
      .sum()
      .sort_values(ascending=False)
)

# ---------------------------------------------------------
# 3. CATEGORY VARIATION (STD DEV)
# ---------------------------------------------------------
variation = (
    df.groupby("Category")["Amount"]
      .agg(["mean", "std", "count"])
      .sort_values("std", ascending=False)
)

# ---------------------------------------------------------
# 4. AI‑ASSISTED BUDGET LOGIC
# ---------------------------------------------------------
budget = monthly_totals.to_frame("Current_Spend")
budget["Suggested_Budget"] = budget["Current_Spend"]

# Rules:
# - High variation → +20% buffer
# - Moderate variation → +10–15%
# - Habit categories (coffee) → +10%
# - Essentials (utilities, transport) → no change
# - Refunds excluded

for category in budget.index:

    spend = budget.loc[category, "Current_Spend"]
    std = variation.loc[category, "std"]

    # Skip refunds
    if category.lower() == "refund":
        budget.loc[category, "Suggested_Budget"] = 0
        continue

    # High variation
    if std > 40:
        budget.loc[category, "Suggested_Budget"] = spend * 1.20

    # Moderate variation
    elif std > 20:
        budget.loc[category, "Suggested_Budget"] = spend * 1.15

    # Habit category
    elif category.lower() == "coffee":
        budget.loc[category, "Suggested_Budget"] = spend * 1.10

    # Essentials (no change)
    elif category.lower() in ["utilities", "transport"]:
        budget.loc[category, "Suggested_Budget"] = spend

    # Default small buffer
    else:
        budget.loc[category, "Suggested_Budget"] = spend * 1.10

# Round for readability
budget["Suggested_Budget"] = budget["Suggested_Budget"].round(2)

# ---------------------------------------------------------
# 5. IDENTIFY REDUCTION OPPORTUNITIES
# ---------------------------------------------------------
budget["Reduction_Potential"] = (
    budget["Current_Spend"] - budget["Suggested_Budget"]
).round(2)

# Positive values = overspending
reduction_targets = budget[budget["Reduction_Potential"] > 0]

# ---------------------------------------------------------
# 6. PACKAGE RESULTS FOR AI NARRATIVE OUTPUT
# ---------------------------------------------------------
ai_summary = {
    "monthly_totals": monthly_totals.to_dict(),
    "variation": variation.to_dict("index"),
    "suggested_budget": budget["Suggested_Budget"].to_dict(),
    "reduction_opportunities": reduction_targets["Reduction_Potential"].to_dict(),
}

ai_summary
````

````python
import pandas as pd

# Load data
df = pd.read_csv("sales_data.csv")

# Rename Region → Store for clarity
df = df.rename(columns={"Region": "Store"})

# Clean numeric columns
for col in ["Quantity", "Unit_Price", "Total_Sale"]:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
        .astype(float)
    )

# -----------------------------
# 1. Store performance summary
# -----------------------------
store_summary = (
    df.groupby("Store")
      .agg(
          Total_Revenue=("Total_Sale", "sum"),
          Total_Units=("Quantity", "sum"),
          Avg_Sale_Value=("Total_Sale", "mean")
      )
      .sort_values("Total_Revenue", ascending=False)
)

print("\n=== Store Performance Summary ===")
print(store_summary)

# -----------------------------
# 2. Product performance
# -----------------------------
product_summary = (
    df.groupby("Product")
      .agg(
          Units_Sold=("Quantity", "sum"),
          Revenue=("Total_Sale", "sum")
      )
      .sort_values("Units_Sold", ascending=False)
)

print("\n=== Product Performance Summary ===")
print(product_summary)

# -----------------------------
# 3. Identify slow-moving products
# -----------------------------
slow_products = product_summary[
    (product_summary["Units_Sold"] < 5) | (product_summary["Revenue"] < 3000)
]

print("\n=== Slow-Moving Products (Discontinue Candidates) ===")
print(slow_products)

# -----------------------------
# 4. Promotion timing (simple logic)
# -----------------------------
df["Date"] = pd.to_datetime(df["Date"])
df["Day"] = df["Date"].dt.day

early_month_sales = df[df["Day"] <= 10]["Total_Sale"].sum()
mid_month_sales = df[(df["Day"] > 10) & (df["Day"] <= 20)]["Total_Sale"].sum()
late_month_sales = df[df["Day"] > 20]["Total_Sale"].sum()

print("\n=== Promotion Timing Insight ===")
print(f"Early-month sales: ${early_month_sales:,.2f}")
print(f"Mid-month sales:   ${mid_month_sales:,.2f}")
print(f"Late-month sales:  ${late_month_sales:,.2f}")

# -----------------------------
# 5. Staffing recommendation
# -----------------------------
staffing = store_summary[["Total_Revenue"]].copy()
staffing["Staffing_Level"] = pd.qcut(
    staffing["Total_Revenue"],
    q=4,
    labels=["Low", "Medium", "High", "Very High"]
)

print("\n=== Staffing Recommendation ===")
print(staffing)
````

AI-assisted personal finance advice generation

````python
import pandas as pd

# -----------------------------
# 1. Load + clean data
# -----------------------------
df = pd.read_csv("transactions.csv")

df["Amount"] = (
    df["Amount"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)

df["Category"] = df["Category"].str.strip()

# -----------------------------
# 2. Typical spending percentages (industry averages)
# -----------------------------
typical = {
    "Groceries": 0.15,
    "Transport": 0.10,
    "Dining": 0.05,
    "Entertainment": 0.05,
    "Coffee": 0.02,
    "Utilities": 0.08,
}

# -----------------------------
# 3. Calculate actual spending percentages
# -----------------------------
total_spent = df[df["Amount"] > 0]["Amount"].sum()

category_totals = (
    df[df["Amount"] > 0]
    .groupby("Category")["Amount"]
    .sum()
    .sort_values(ascending=False)
)

actual_pct = (category_totals / total_spent).round(3)

# -----------------------------
# 4. Identify overspending categories
# -----------------------------
overspending = {}

for cat, pct in actual_pct.items():
    if cat in typical and pct > typical[cat]:
        overspending[cat] = {
            "actual": pct,
            "typical": typical[cat],
            "difference": round(pct - typical[cat], 3)
        }

# -----------------------------
# 5. Suggest savings opportunities
# -----------------------------
savings_suggestions = []

for cat, info in overspending.items():
    if info["difference"] > 0.03:
        savings_suggestions.append(
            f"You’re spending more than usual on **{cat}**. "
            f"Try reducing this category by 5–10% to stay on track."
        )
    else:
        savings_suggestions.append(
            f"Your **{cat}** spending is slightly above average. "
            f"A small adjustment could help balance your budget."
        )

# -----------------------------
# 6. Encouraging feedback
# -----------------------------
positive_feedback = []

for cat, pct in actual_pct.items():
    if cat in typical and pct <= typical[cat]:
        positive_feedback.append(
            f"Great job keeping **{cat}** spending under control!"
        )

# -----------------------------
# 7. Format user-friendly advice
# -----------------------------
print("\n=== Your Spending Advice ===\n")

print("📌 Categories Above Typical Levels:")
if overspending:
    for cat, info in overspending.items():
        print(f"- {cat}: {info['actual']*100:.1f}% (typical {info['typical']*100:.1f}%)")
else:
    print("No categories above typical spending — well done!")

print("\n💰 Suggested Savings Areas:")
if savings_suggestions:
    for s in savings_suggestions:
        print(f"- {s}")
else:
    print("You're spending wisely across all categories.")

print("\n🌟 Positive Feedback:")
for p in positive_feedback:
    print(f"- {p}")

print("\nTotal Spending:", f"${total_spent:,.2f}")
````
