```python
# Week 8: My Data Story
# This notebook reads and analyzes the "my_data.txt" file created earlier.

# Step 1: Open and read the file
with open("my_data.txt", "r") as file:
    content = file.readlines()  # reads the file into a list, one line per element

# Step 2: Display the file contents clearly
print("=== File Contents ===\n")
for line in content:
    print(line.strip())

# Step 3: Analyze the file
# Calculate number of lines, words, total characters, and longest line
num_lines = len(content)
num_words = sum(len(line.split()) for line in content)
char_count = sum(len(line) for line in content)
longest_line = max(content, key=len).strip()

# Step 4: Display the results clearly
print("\n=== File Analysis ===")
print(f"Number of lines: {num_lines}")
print(f"Number of words: {num_words}")
print(f"Total character count: {char_count}")
print(f"Longest line: {longest_line}")
```
