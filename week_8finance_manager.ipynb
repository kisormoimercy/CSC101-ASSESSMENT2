```python
def load_transactions(filename):
    """
    Load transaction lines from a file safely.
    Returns a cleaned list of non‑empty lines.
    """

    try:
        # Try to open the file in read‑only mode.
        # If the file does not exist, this line will raise FileNotFoundError.
        with open(filename, 'r') as file:
            lines = file.readlines()   # Read all lines into a list

        # Clean each line by removing whitespace/newlines.
        # The 'if line.strip()' part filters out any empty or blank lines.
        cleaned = [line.strip() for line in lines if line.strip()]
        return cleaned

    except FileNotFoundError:
        # This block runs if the file cannot be found.
        # Instead of crashing, we show a friendly message and return an empty list.
        print(f"File '{filename}' not found!")
        return []
```

```python
def load_transactions(filename):
    """Reads all lines from the file safely."""
    try:
        with open(filename, "r") as f:
            lines = f.readlines()
        return lines
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
        return []


def process_transactions(lines):
    """Cleans and validates the transaction lines."""
    transactions = []
    failed_count = 0

    for line in lines:
        line = line.strip()
        if not line or ',' not in line:
            failed_count += 1
            continue

        parts = line.split(',')
        if len(parts) != 2:
            failed_count += 1
            continue

        description, amount_text = parts
        description = description.strip()

        try:
            amount = float(amount_text.strip())
        except ValueError:
            failed_count += 1
            continue

        transactions.append({
            "description": description,
            "amount": amount
        })

    print(f"{failed_count} lines could not be processed.")
    return transactions


def summarize_transactions(transactions):
    """Shows totals for valid transactions."""
    total_transactions = len(transactions)
    total_income = sum(t["amount"] for t in transactions if t["amount"] > 0)
    total_expenses = sum(t["amount"] for t in transactions if t["amount"] < 0)
    net_total = total_income + total_expenses

    print("\n==== TRANSACTION SUMMARY ====")
    print(f"Total transactions processed: {total_transactions}")
    print(f"Total income:   ${total_income:.2f}")
    print(f"Total expenses: ${total_expenses:.2f}")
    print(f"Net amount:     ${net_total:.2f}")
    print("=============================\n")


# 🚀 MAIN PROGRAM
filename = "transactions.txt"
raw_lines = load_transactions(filename)
transactions = process_transactions(raw_lines)
summarize_transactions(transactions)
```
